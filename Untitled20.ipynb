```python
import pandas as pd
```

```python
df = pd.read_csv('/content/ICRISAT-District Level Data.csv')
display(df.head())
```

```python
df.info()
```

```python
display(df.describe())
```

```python
import numpy as np

# Identify columns with -1 values
minus_one_cols = df.columns[(df == -1).any()].tolist()

print("Columns containing -1.0 values:")
for col in minus_one_cols:
    count = (df[col] == -1).sum()
    print(f"- {col}: {count} occurrences")
```

```python
for col in minus_one_cols:
    df[col] = df[col].replace(-1, np.nan)

print("Number of NaN values per column after replacing -1:")
# Display the count of NaN values for each column containing -1.0 values
for col in minus_one_cols:
    print(f"- {col}: {df[col].isna().sum()} NaN values")
```

```python
rice_prod = df.groupby('State Name')['RICE PRODUCTION (1000 tons)'].sum()

top_rice = rice_prod.sort_values(ascending=False).head(10)

print(top_rice)
```

```python
import matplotlib.pyplot as plt

top_rice.plot(kind='bar')

plt.title('Top Rice Producing States')
plt.xlabel('States')
plt.ylabel('Production')

plt.show()
```

```python
corr_cols = [
    'RICE AREA (1000 ha)',
    'RICE PRODUCTION (1000 tons)',
    'RICE YIELD (Kg per ha)'
]

corr = df[corr_cols].corr()

print(corr)
```

```python
print("Total NaN values in the DataFrame after imputation:", df.isnull().sum().sum())
```

```python
print("Unique values and their counts for 'State Name':")
display(df['State Name'].value_counts())

print("\nUnique values and their counts for 'Dist Name':")
display(df['Dist Name'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

yearly_rice_production = df.groupby('Year')['RICE PRODUCTION (1000 tons)'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=yearly_rice_production, x='Year', y='RICE PRODUCTION (1000 tons)')
plt.title('Total Rice Production Over the Years (All India)')
plt.xlabel('Year')
plt.ylabel('Total Rice Production (1000 tons)')
plt.grid(True)
plt.show()
```

```python
top_states_rice = df.groupby('State Name')['RICE PRODUCTION (1000 tons)'].sum().nlargest(10).reset_index()

plt.figure(figsize=(14, 7))
sns.barplot(x='RICE PRODUCTION (1000 tons)', y='State Name', data=top_states_rice, palette='viridis')
plt.title('Top 10 States by Total Rice Production (1000 tons)')
plt.xlabel('Total Rice Production (1000 tons)')
plt.ylabel('State Name')
plt.grid(axis='x')
plt.show()
```

```python
top_states_names = top_states_rice['State Name'].tolist()
df_top_states = df[df['State Name'].isin(top_states_names)]

plt.figure(figsize=(15, 8))
sns.lineplot(data=df_top_states, x='Year', y='RICE PRODUCTION (1000 tons)', hue='State Name', marker='o')
plt.title('Rice Production Trends in Top 10 States Over the Years')
plt.xlabel('Year')
plt.ylabel('Rice Production (1000 tons)')
plt.grid(True)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
correlation_cols = [
    'RICE PRODUCTION (1000 tons)', 'RICE AREA (1000 ha)', 'RICE YIELD (Kg per ha)',
    'WHEAT PRODUCTION (1000 tons)', 'WHEAT AREA (1000 ha)', 'WHEAT YIELD (Kg per ha)',
    'MAIZE PRODUCTION (1000 tons)', 'MAIZE AREA (1000 ha)', 'MAIZE YIELD (Kg per ha)',
    'SUGARCANE PRODUCTION (1000 tons)', 'SUGARCANE AREA (1000 ha)', 'SUGARCANE YIELD (Kg per ha)'
]

# Calculate the correlation matrix
corr_matrix = df[correlation_cols].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Key Agricultural Metrics')
plt.show()
```

```python
rice_production_correlations = corr_matrix['RICE PRODUCTION (1000 tons)'].sort_values(ascending=False)

# Exclude the correlation of 'RICE PRODUCTION (1000 tons)' with itself
rice_production_correlations = rice_production_correlations[rice_production_correlations.index != 'RICE PRODUCTION (1000 tons)']

display(rice_production_correlations.head(10))
```

From the correlations, `RICE AREA (1000 ha)` and `RICE YIELD (Kg per ha)` show the highest positive correlation with rice production, which is expected. Among other crops, `MAIZE PRODUCTION (1000 tons)`, `SUGARCANE PRODUCTION (1000 tons)`, and `WHEAT PRODUCTION (1000 tons)` also show positive correlations, albeit weaker ones.
