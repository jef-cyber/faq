### Data Preprocessing: Encoding Categorical Features

Before encoding other categorical features, it's good practice to convert any date columns to datetime objects if they are not already. The 'Date' column is currently an `object` type, so we'll convert it first.

```python
df['Date'] = pd.to_datetime(df['Date'])
display(df.info())
```

For binary categorical columns (like 'Spoilage_Observed' and 'Storage_Available' which seem to have 'Yes'/'No' values), `LabelEncoder` is a simple and effective method. It assigns a unique integer to each unique category.

Let's apply `LabelEncoder` to 'Spoilage_Observed' and 'Storage_Available'.

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
le = LabelEncoder()

# Apply Label Encoding to 'Spoilage_Observed'
df['Spoilage_Observed_Encoded'] = le.fit_transform(df['Spoilage_Observed'])

# Apply Label Encoding to 'Storage_Available'
df['Storage_Available_Encoded'] = le.fit_transform(df['Storage_Available'])

# Display the updated DataFrame with new encoded columns and their first few rows
display(df[['Spoilage_Observed', 'Spoilage_Observed_Encoded', 'Storage_Available', 'Storage_Available_Encoded']].head())
```

For nominal categorical columns with multiple unique values (like 'Location' and 'Crop_Type'), **One-Hot Encoding** is generally preferred. This method creates new binary columns for each category, preventing the model from assuming an ordinal relationship between categories.

Let's apply One-Hot Encoding to 'Location' and 'Crop_Type'.

```python
df_encoded = pd.get_dummies(df, columns=['Location', 'Crop_Type'], drop_first=True)

# Display the updated DataFrame with one-hot encoded columns
display(df_encoded.head())
```
